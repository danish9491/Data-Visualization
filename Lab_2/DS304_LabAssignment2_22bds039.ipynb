# Q2_a: Exploring Univariate Data Visualization



```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Loading the mtcars dataset
mtcars = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/mtcars.csv')

# Display the first few rows of the dataset to understand its structure
mtcars.head()
```

```python
# Plotting the distribution of the mpg variable
plt.figure(figsize=(10, 6))
sns.histplot(mtcars['mpg'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of MPG')
plt.xlabel('Miles per Gallon (mpg)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plotting the counts of cars for each cylinder type
plt.figure(figsize=(10, 6))
sns.countplot(x='cyl', data=mtcars, palette='viridis')
plt.title('Counts of Cars for Each Cylinder Type')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

```python
# Plotting the proportion of cars with automatic vs. manual transmission
plt.figure(figsize=(8, 8))
am_counts = mtcars['am'].value_counts()
plt.pie(am_counts, labels=['Automatic', 'Manual'], autopct='%1.1f%%', colors=['lightcoral', 'lightblue'], startangle=140)
plt.title('Proportion of Cars with Automatic vs. Manual Transmission')
plt.show()
```

```python
# Plotting the distribution of horsepower using a boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(x=mtcars['hp'], color='lightgreen')
plt.title('Boxplot of Horsepower (hp)')
plt.xlabel('Horsepower')
plt.show()
```

```python
# Plotting the frequency distribution of cars with different numbers of gears
plt.figure(figsize=(10, 6))
sns.countplot(x='gear', data=mtcars, palette='pastel')
plt.title('Frequency Distribution of Cars with Different Numbers of Gears')
plt.xlabel('Number of Gears')
plt.ylabel('Count')
plt.show()
```

```python
# Scatter plot showing the relationship between mpg and hp
plt.figure(figsize=(10, 6))
sns.scatterplot(x='hp', y='mpg', data=mtcars, color='purple', s=100)
plt.title('Scatter Plot of MPG vs. Horsepower')
plt.xlabel('Horsepower (hp)')
plt.ylabel('Miles per Gallon (mpg)')
plt.show()
```

```python
# Scatter plot showing the relationship between weight and mpg
plt.figure(figsize=(10, 6))
sns.regplot(x='wt', y='mpg', data=mtcars, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
plt.title('Scatter Plot of Weight vs. MPG with Regression Line')
plt.xlabel('Weight (1000 lbs)')
plt.ylabel('Miles per Gallon (mpg)')
plt.show()
```

# Q2_b: Visualize Datasaurus dozen

```python
# Load the Datasaurus Dozen dataset from the TSV file
df = pd.read_csv('DatasaurusDozen.tsv', sep='\t')
```

```python
df
```

```python
# Display the first few rows of the dataset to understand its structure
print(df.head(), "\n")
```

```python
# Get the unique dataset names
datasets = df['dataset'].unique()
```

```python
datasets
```

```python
# Calculate the number of rows and columns for the subplot grid
import math
n_datasets = len(datasets)
n_cols = math.ceil(math.sqrt(n_datasets))  # Columns based on square root of total datasets
n_rows = math.ceil(n_datasets / n_cols)  # Rows to fit all datasets
```

```python
print("Number of datasets:", n_datasets)
```

```python
print("Number of columns:", n_cols)
```

```python
print("Number of rows:", n_rows)
```

```python
# Create a figure to plot all datasets
plt.figure(figsize=(n_cols * 5, n_rows * 5))

# Loop over each unique dataset and create a scatter plot
for i, dataset in enumerate(datasets):
    plt.subplot(n_rows, n_cols, i + 1)
    subset = df[df['dataset'] == dataset]
    plt.scatter(subset['x'], subset['y'], color='dodgerblue')
    plt.title(dataset)
    plt.xlim(0, 100)
    plt.ylim(0, 100)
    plt.xlabel('x')
    plt.ylabel('y')

# Add a main title to the entire figure
plt.suptitle('Datasaurus Dozen Visualizations', fontsize=22, y=1.02)
plt.tight_layout()
plt.show()
```
